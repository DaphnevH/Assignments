## Graphs

To make the graphs we need to make selections of the data. Please use "dplyr" (commands "select" and "filter") and the "pipe". The first assignment is to make a scatterplot using "ggplot". The graph contains data for all countries for the year 2014. Therefore make a dataframe with dplyr first, called "graph1" with the required data.
Then, make a scatterplot with "Cost" on the x-axis and "LE" on the y-axis. You have to add a regression line with a shaded area for the confidence interval.

**Uitleg**:
* _Select_ = Door gebruik te maken van select() kun je variabelen selecteren. 
* _Filter_ = Door gebruik te maken van de functie filter() kun je een 'logical test' doorgeven waarop gefilterd wordt.

**Explanation**:
* _Select_ = With the command select(), variables can be selected 
--> In this case, we select all the X's, countries, cost and LE.
* _Filter_ = With the command filter(), a 'logical test' can be made and will be filtered 
--> In this case is the logical test that the information must come from the year 2014 (Year == 2014) and all that data is shown.
* _Pipe_ = With %>%, R uses the command in the previous line for the next line 
--> R works further with the fact that graph1 the new name is for oecd.


```r
graph1 <-oecd %>%
    filter(Year == 2014) %>%
    select(X, Country, Cost, LE)
```

```r
ggplot(graph1, aes(x = Cost, y = LE, label = Country)) + geom_point() + geom_smooth(method=lm) + geom_text()
```

## Efficiency scores

The last assignment is to compute simple effiency scores. We will use the year 2014 for all countries (i.e. the dataframe "graph1").
Make a new dateframe called "eff" (using graph1). Please use the "pipe" to execute to following steps:
Select all the variables except the NA's (hint: use na.omit) and create a new column called "Ratio" with mutate. Ratio is defined as "Cost/LE" (cost per output).
Create a new column called "Score" with mutate. Score is the ratio per country divided by the lowest ratio in the dataframe.
To complete this dataframe order the dataset using "arrange" based on score (i.e. lowest Score first).


Finally make a bar chart using ggplot. Countries should be on the x-axis and "Score" on the y-axis. 



You can earn extra points if you are able to order the graph based on scores (lowest first, highest last), if you are able to make the x-axis readable, make the bar of the Netherlands red and the other bars blue and/or change the name of the y-axis to "Efficiency scores according to [your name]" 


The figure should like this:

![My Figure](eff.png)

**Explanation**: 
A bar chart is a graph with bars for every country. It can be very easy to use if you want to compare different countries in a quick way. This bar chart is made from low to high. You can see things very quick in this way. 
Na.omit() is used because with this command it is possible to delete al the information from incomplete obervations. The command mutate() helps you to change the data that is available. In this graph, we make an extra collumn named Ratio for example. 


```r
eff <- graph1 [,c(0,1,2,3,4)] %>%
    na.omit(eff) %>%
    mutate(Ratio = Cost / LE) %>%
    mutate(Score = min(Ratio) / Ratio) %>%
    arrange(Score, Country)

eff$col <- ifelse(eff$Country == "Netherlands", "red", "blue")
```

```r
ggplot(eff, aes(x = reorder(Country, Score), y=Score, fill=col)) + geom_bar(stat="identity") + scale_fill_identity() + theme(axis.text.x=element_text(angle=75,hjust=1,vjust=1)) + labs(x="Country", y="Efficiency score according to Daphne")
```

**Explanation**: The graph above is the expected graph. As you can see, United States has the lowest efficiency score and Turkey the highest.
